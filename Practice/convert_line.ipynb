```python
df1 = df.loc[:,'year':'id']
df1.dropna()
```

' . '을 기준으로 분리해준다 (문장단위)

```python
split = df.document.str.split('.')
split
```

```python
split = split.apply(lambda x: pd.Series(x))
split
```

```python
split.stack()
```

```python
split.stack().reset_index(level=1, drop=True)
```

```python
split = split.stack().reset_index(level=1, drop=True).to_frame('document')
split
```

총 31368개의 행으로 이루어진 데이터프레임을 얻을 수 있다.



```python
merge = pd.merge(df1,split, how='outer', left_index=True, right_index=True)
merge_df = merge.dropna()
merge_df
```
